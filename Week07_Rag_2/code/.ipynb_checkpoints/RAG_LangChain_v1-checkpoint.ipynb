## Loading and Splitting

```python
!pip install -qU langchain langchain-community pypdf
```

```python
from langchain_community.document_loaders import PyPDFLoader
```

```python
# Bulk Loading PDF
loaders = [
    PyPDFLoader("MachineLearning-Lecture01.pdf"),
    PyPDFLoader("MachineLearning-Lecture01.pdf"),
    PyPDFLoader("MachineLearning-Lecture02.pdf"),    
    PyPDFLoader("MachineLearning-Lecture03.pdf"),    
]
docs = []
for loader in loaders:
    docs.extend(loader.load())
```

```python
# Split
!pip install -qU langchain-text-splitters

from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(
    chunk_size = 1500,
    chunk_overlap = 150
)
```

```python
splits = text_splitter.split_documents(docs)
```

```python
len(splits)
```

```python
from langchain_openai import OpenAIEmbeddings
from dotenv import load_dotenv

load_dotenv()

embedding = OpenAIEmbeddings(
    model="text-embedding-3-large"
)
```

## Vectorstores

```python
!pip install -qU chromadb langchain-chroma
```

```python
persist_directory = 'docs/chroma'
!rm -rf ./docs/chroma
```

```python
from langchain_chroma import Chroma
from langchain_openai import OpenAIEmbeddings

vectordb = Chroma.from_documents(
    documents=splits,
    embedding=embedding,
    persist_directory=persist_directory
)
```

```python
print(vectordb._collection.count())
```

## Similarity Search

```python
question = "is there an email i can ask for help"
docs = vectordb.similarity_search(question, k=3)
```

#### Agentic RAG using Tools

```python
from langchain.tools import tool

@tool(response_format="content_and_artifact")
def retrieve_context(query: str):
    """Retrieve information to help answer a query."""
    retrieved_docs = vectordb.similarity_search(query, k=2)
    serialized = "\n\n".join(
        (f"Source: {doc.metadata}\nContent: {doc.page_content}")
        for doc in retrieved_docs
    )
    return serialized, retrieved_docs
```

```python
!pip install -qU dotenv
!pip install -qU langchain-openai    
```

```python
load_dotenv()
```

```python
from langchain.chat_models import init_chat_model

model = init_chat_model(
    "openai:gpt-4o",
    temperature=0.5,
    timeout=10,
    max_tokens=1000
)
```

```python
from langchain.agents import create_agent

tools = [retrieve_context]
# If desired, specify custom instructions
prompt = (
    "You have access to a tool that retrieves context from course transcript"
    "Use the tool to help answer user queries."
)
agent = create_agent(model, tools, system_prompt=prompt)
```

#### 3 Ways to Invoke an Agent
1. invoke() -- single call
2. stream() -- stream output while being generated
3. batch() -- clubbing multiple independant request - reduce cost / improves performance

```python
# Invoking the agent
response = agent.invoke({"messages": [{"role": "user", "content": user_query}]})
```
