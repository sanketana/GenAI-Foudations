# Section 2: Vector Operations

```python
import numpy as np
```

```python
vector_a = np.array([3, 4])
vector_b = np.array([1, 2])
```

```python
print(vector_a, vector_b)
```

```python
dot_product = np.dot(vector_a, vector_b)
print(dot_product)
```

```python
# Vectors are parallel (very similar)
parallel_a = np.array([2, 0])
parallel_b = np.array([4, 0])

dot_product = np.dot(parallel_a, parallel_b)
print(dot_product)
```

```python
# Vectors are perpendicular (NOT similar)
parallel_c = np.array([0, 3])

dot_product = np.dot(parallel_c, parallel_b)
print(dot_product)
```

```python
# Dot product of multidimensional vector
parallel_a = np.array([2, 0, 0])
parallel_b = np.array([0, 4, 5])

dot_product = np.dot(parallel_a, parallel_b)
print(dot_product)
```
